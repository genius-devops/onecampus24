```python
# 4. write a function that can fetch a students score for each subject

def get_student_scores(student_name, subjects, student_names, subject_scores):
    scores = {}
    
    # Find the index of the student in the list of student names
    student_index = student_names.index(student_name) if student_name in student_names else None
    
    # If the student is found, retrieve their scores for each subject
    if student_index is not None:
        for subject, scores_list in zip(subjects, subject_scores):
            # Check if the subject index is within the range of the scores list
            if student_index < len(scores_list):
                scores[subject] = scores_list[student_index]
            else:
                scores[subject] = None
    else:
        print(f"Student '{student_name}' not found.")
    
    return scores

# Example usage:
student_name = 'Kelechi'
student_scores = get_student_scores(student_name, subjectList, chemStuds, [chemScores, bioScores, mathScores, phyScores])

print(f"{student_name}'s scores:")
for subject, score in student_scores.items():
    print(f"{subject}: {score}")
```

```python
# 5. write function to grade the student scores for each subject and record it into a new list             

def grade_student_scores(scores_list):
    graded_scores = []

    for scores in scores_list:
        # Define grading criteria based on your requirements
        if scores >= 90:
            grade = 'A'
        elif 80 <= scores < 90:
            grade = 'B'
        elif 70 <= scores < 80:
            grade = 'C'
        elif 60 <= scores < 70:
            grade = 'D'
        else:
            grade = 'F'

        graded_scores.append(grade)

    return graded_scores

# Example usage:
chem_grades = grade_student_scores(chemScores)
bio_grades = grade_student_scores(bioScores)
math_grades = grade_student_scores(mathScores)
phy_grades = grade_student_scores(phyScores)

print("Chemistry Grades:", chem_grades)
print("Biology Grades:", bio_grades)
print("Math Grades:", math_grades)
print("Physics Grades:", phy_grades)
```

```python
# 6. Write code to store all the student results and grades obtained from Q5 in a single dictionary  

def get_student_results_and_grades(subjects, student_names, subject_scores):
    results_and_grades = {}

    for subject, scores_list in zip(subjects, subject_scores):
        subject_results = {}
        for student_name, score in zip(student_names, scores_list):
            # Determine the grade based on your grading criteria
            if score >= 90:
                grade = 'A'
            elif 80 <= score < 90:
                grade = 'B'
            elif 70 <= score < 80:
                grade = 'C'
            elif 60 <= score < 70:
                grade = 'D'
            else:
                grade = 'F'

            subject_results[student_name] = {'Score': score, 'Grade': grade}

        results_and_grades[subject] = subject_results

    return results_and_grades

# Example usage:
all_results_and_grades = get_student_results_and_grades(subjectList, chemStuds, [chemScores, bioScores, mathScores, phyScores])

# Print the dictionary
for subject, results in all_results_and_grades.items():
    print(f"\n{subject} Results and Grades:")
    for student_name, data in results.items():
        print(f"{student_name}: Score - {data['Score']}, Grade - {data['Grade']}")
```

```python
# 7. Write a single function that can be used to do the following
             
             
#              1. find all the subjects a student offers
#              2. fetch a students' exam score for a subject or all the subject he/she offers
#              3. grade the exam scores of students and store the information in a dictionary
#              4. input subject exam scores for students
#              5. change or update the exam scores for any subject for a given student
#              6. find all the students whose grade or score is below a specified value or within a specified range

def student_management_system(subjects, student_names, subject_scores, grading_criteria):
    """
    Student Management System Function

    Args:
    - subjects: List of subjects offered
    - student_names: List of student names
    - subject_scores: List of lists containing exam scores for each subject and student
    - grading_criteria: Dictionary specifying grading criteria for each subject

    Returns:
    - A dictionary containing information about subjects, student scores, grades, and related operations.
    """

    data = {
        'subjects': subjects,
        'students': {},
    }

    # Function to find all the subjects a student offers
    def find_student_subjects(student_name):
        student_index = student_names.index(student_name) if student_name in student_names else None
        if student_index is not None:
            return [subject for subject, scores_list in zip(subjects, subject_scores) if student_index < len(scores_list)]
        else:
            return None

    # Function to fetch a student's exam score for a subject or all subjects
    def get_student_scores(student_name, selected_subject=None):
        student_index = student_names.index(student_name) if student_name in student_names else None
        if student_index is not None:
            scores_dict = {}
            for subject, scores_list in zip(subjects, subject_scores):
                if selected_subject is None or selected_subject == subject:
                    scores_dict[subject] = scores_list[student_index]
            return scores_dict
        else:
            return None

    # Function to grade the exam scores of students and store the information in a dictionary
    def grade_exam_scores():
        graded_scores = {}
        for subject, scores_list, criteria in zip(subjects, subject_scores, grading_criteria):
            graded_scores[subject] = [criteria(score) for score in scores_list]
        return graded_scores

    # Function to input subject exam scores for students
    def input_exam_scores(student_name, subject, score):
        student_index = student_names.index(student_name) if student_name in student_names else None
        if student_index is not None:
            subject_index = subjects.index(subject) if subject in subjects else None
            if subject_index is not None:
                subject_scores[subject_index][student_index] = score
                return True
        return False

    # Function to change or update the exam scores for any subject for a given student
    def update_exam_scores(student_name, subject, new_score):
        return input_exam_scores(student_name, subject, new_score)

    # Function to find all the students whose grade or score is below a specified value or within a specified range
    def find_students_by_grade(grade_threshold, subject=None, score_range=None):
        filtered_students = {}
        for student_name in student_names:
            student_scores = get_student_scores(student_name, subject)
            if student_scores:
                student_grade = grading_criteria(subject_scores[subjects.index(subject)][student_names.index(student_name)])
                if student_grade < grade_threshold:
                    filtered_students[student_name] = {'Scores': student_scores, 'Grade': student_grade}
                elif score_range and score_range[0] <= student_scores[subject] <= score_range[1]:
                    filtered_students[student_name] = {'Scores': student_scores, 'Grade': student_grade}
        return filtered_students

    # Populate the 'students' dictionary with initial data
    for student_name in student_names:
        data['students'][student_name] = {'Subjects': find_student_subjects(student_name),
                                          'Scores': get_student_scores(student_name),
                                          'Grades': grading_criteria(subject_scores[0][student_names.index(student_name)])}

    # Include functions in the data dictionary for external access
    data['find_student_subjects'] = find_student_subjects
    data['get_student_scores'] = get_student_scores
    data['grade_exam_scores'] = grade_exam_scores
    data['input_exam_scores'] = input_exam_scores
    data['update_exam_scores'] = update_exam_scores
    data['find_students_by_grade'] = find_students_by_grade

    return data

# Example usage:
grading_criteria = lambda score: 'A' if score >= 90 else ('B' if 80 <= score < 90 else ('C' if 70 <= score < 80 else 'F'))

sms_data = student_management_system(subjectList, chemStuds, [chemScores, bioScores, mathScores, phyScores], [grading_criteria])

# Example function calls:
print(sms_data['get_student_scores']('Kelechi', 'chem'))
print(sms_data['grade_exam_scores']())
sms_data['input_exam_scores']('Kelechi', 'chem', 95)
print(sms_data['get_student_scores']('Kelechi', 'chem'))
sms_data['update_exam_scores']('Kelechi', 'chem', 80)
print(sms_data['get_student_scores']('Kelechi', 'chem'))
print(sms_data['find_students_by_grade'](80, 'chem'))
```
